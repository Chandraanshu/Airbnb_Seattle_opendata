```python
####SECTION 4: CLEAN UP OF DATA
data.host_response_time.unique()
```

```python
#Change host response time categories to numbers
data['host_response_time'].replace('within an hour', 0, inplace=True)
data['host_response_time'].replace('within a few hours', 0.25, inplace=True)
data['host_response_time'].replace('within a day', 0.5, inplace=True)
data['host_response_time'].replace('a few days or more', 0.75, inplace=True)
data['host_response_time'].fillna(method = 'ffill', inplace = True)
```

```python
#Changing price to numbers
data['cleaning_fee'] = data['cleaning_fee'].str.replace('$', '')
data['cleaning_fee'] = data['cleaning_fee'].str.replace(',', '').astype('float64')
data['cleaning_fee'].fillna(0, inplace=True)

data['price'] = data['price'].str.replace('$', '')
data['price'] = data['price'].str.replace(',', '').astype('float64')
data['extra_people'] = data['extra_people'].str.replace('$', '')
data['extra_people'] = data['extra_people'].str.replace(',', '').astype('float64')
```

```python
#Change True/False to one-hot binary
data['host_is_superhost'].replace('t',1, inplace=True)
data['host_is_superhost'].replace('f',0, inplace=True)
data['instant_bookable'].replace('f',0, inplace=True)
data['instant_bookable'].replace('t',1, inplace=True)
```

```python
#Binary property type to whether it is proper house or not
def is_proper_house(i):
    if i in ['House', 'Apartment', 'Townhouse', 'Condominium', 'Bungalow', 'Loft']:
        return 1
    else :
        return 0

data['is_proper_house'] = data['property_type'].apply(is_proper_house)

#Binary property type to whether has proper bed or not
def is_proper_bed(i):
    if i == 'Real Bed':
        return 1
    else :
        return 0
data['is_proper_bed'] = data['bed_type'].apply(is_proper_bed)

#Binary to see if room is private or shared
def is_private_room(i):
    if i in ['Entire home/apt', 'Private room']:
        return 1
    else :
        return 0
data['is_private_room'] = data['room_type'].apply(is_private_room)

data.drop(['property_type', 'bed_type', 'room_type'], axis = 1, inplace = True)
```

```python
#Change rates to numbers
data['host_response_rate'] = data['host_response_rate'].replace('%','',regex=True)
data['host_acceptance_rate'] = data['host_acceptance_rate'].replace('%','',regex=True)

data['host_response_rate'] = data['host_response_rate'].astype('float64') / 100
data['host_acceptance_rate'] = data['host_acceptance_rate'].astype('float64') / 100

data['host_response_rate'].fillna(method='ffill', inplace=True)
data['host_acceptance_rate'].fillna(method='ffill', inplace=True)
```

```python
data.beds.isna().sum() + data.bathrooms.isna().sum() + data.bedrooms.isna().sum() + data.host_is_superhost.isna().sum() 
```

```python
#Dropping the 25 rows with NA values
data.dropna(subset=['beds', 'bathrooms', 'bedrooms', 'host_is_superhost'], inplace = True)
```

```python
data.review_scores_rating.isna().sum()
```

```python
#Dropping the rows with NA values for our prediction variable
data.dropna(subset=['review_scores_rating'], inplace = True)
```

```python
#Changing amenities to number of amenities
data['number_amenities'] = data['amenities'].str.count(',') + 1
data.drop(columns = ['amenities'], inplace = True)
```

```python
#Adding sentiment analysis of descriptions
from vaderSentiment.vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sentiment_analyzer = SentimentIntensityAnalyzer()

def get_sentiment(i):
    sentiment = sentiment_analyzer.polarity_scores(i)
    return sentiment.get('pos')

data['description_sentiment'] = data['description'].apply(get_sentiment)
data['description_length'] = data['description'].str.len()

data['name_sentiment'] = data['name'].apply(get_sentiment)
data['name_length'] = data['name'].str.len()

data.drop(columns = ['description', 'name'], inplace = True)
```

```python
####SECTION 6: TRAINING REGRESSION MODELS

data_training = data.loc[:,('host_is_superhost', 'accommodates', 'bathrooms', 
                            'bedrooms', 'beds', 'price', 'cleaning_fee', 'instant_bookable',
                            'is_proper_house', 'is_proper_bed', 'is_private_room','number_amenities', 'description_length', 'review_scores_rating')]
data_training.shape
```

```python
#Split into training and testing sets
import sklearn
from sklearn.model_selection import train_test_split
X = data_training.drop(['review_scores_rating'], axis = 1)
y = data_training['review_scores_rating']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 41)
```

```python
#Train linear regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression() 
lin_reg.fit(X_train, y_train) 
```

```python
from sklearn import metrics
lin_reg_mse = metrics.mean_squared_error(y_test, lin_reg.predict(X_test))
lin_reg_mse = np.sqrt(lin_reg_mse)
lin_reg_mse
```

```python
#Train decision tree
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor() 
tree_reg.fit(X_train, y_train) 
```

```python
des_tree_mse = metrics.mean_squared_error(y_test, tree_reg.predict(X_test))
des_tree_mse = np.sqrt(des_tree_mse)
des_tree_mse
```

```python
#Improving decision tree hyperparameters with GridSearchCV
from sklearn.model_selection import GridSearchCV
num_leafs = [2, 5, 10, 20, 50, 100]
param_grid = [{'max_leaf_nodes':num_leafs}]

tree_reg = DecisionTreeRegressor() 
grid_search = GridSearchCV(tree_reg, param_grid, cv=5)
grid_search.fit(X_train, y_train)

tree_reg.fit(X_train, y_train) 
des_tree_mse = metrics.mean_squared_error(y_test, tree_reg.predict(X_test))
des_tree_mse = np.sqrt(des_tree_mse)
des_tree_mse
```

```python
#Train random forest regressor
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(X_train, y_train) 
```

```python
ran_for_mse = metrics.mean_squared_error(y_test, forest_reg.predict(X_test))
ran_for_mse = np.sqrt(ran_for_mse)
ran_for_mse
```

```python
#Improving forest regressor hyperparameters with GridSearchCV
param_grid = [{'n_estimators': [3, 10, 30], 
               'max_features': [2, 4, 6, 8]},
              {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]}]
forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
grid_search.best_params_ 
```

```python
forest_reg.fit(X_train, y_train) 
ran_for_mse = metrics.mean_squared_error(y_test, forest_reg.predict(X_test))
ran_for_mse = np.sqrt(ran_for_mse)
ran_for_mse
```
